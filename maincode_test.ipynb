```python
import numpy as np
import os
data= np.loadtxt(open("Example.csv","rb"),delimiter=",",skiprows=0)
data
```

```python
m,n=data.shape
```

```python
print(m)
print(n)
# print(len(data[0]))
# print(len(data))
```

```python
#行好友的索引
listA=[]
for i in range(len(data)):
    listTemp=[]
    for j in range(len(data[0])):
        if(data[i][j]==1):
            listTemp.append(j+m)
    listA.append(listTemp)
```

```python
listA
```

```python
# 列的好友索引
listB=[]
for j in range(len(data[0])):
    listTemp=[]
    for i in range(len(data)):
        if(data[i][j]==1):
            listTemp.append(i)
    listB.append(listTemp)
```

```python
listB
```

```python
total=listA+listB
```
